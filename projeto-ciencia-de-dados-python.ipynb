### **3.3. Dados faltantes** 

Dados faltantes podem ser:

 - Vazios (`""`);
 - Nulos (`None`);
 - Não disponíveis ou aplicaveis (`na`, `NA`, etc.);
 - Não numérico (`nan`, `NaN`, `NAN`, etc).

```python
df.head()
```

Podemos verificar quais colunas possuem dados faltantes.

```python
df.isna().any()
```

 - A função abaixo levanta algumas estatisticas sobre as colunas dos dados faltantes.

```python
def stats_dados_faltantes(df: pd.DataFrame) -> None:

  stats_dados_faltantes = []
  for col in df.columns:
    if df[col].isna().any():
      qtd, _ = df[df[col].isna()].shape
      total, _ = df.shape
      dict_dados_faltantes = {col: {'quantidade': qtd, "porcentagem": round(100 * qtd/total, 2)}}
      stats_dados_faltantes.append(dict_dados_faltantes)

  for stat in stats_dados_faltantes:
    print(stat)
```

```python
stats_dados_faltantes(df=df)
```

Como observamos que a base de dados é desbalanceada, vamos aplicar esse mesmo método de contagem mas com o filtro de default = 0 e 1, para assim entendermos se esses dados faltantes estão proporcionais tanto para adimplentes quanto para inadimplentes.

```python
stats_dados_faltantes(df=df[df['default'] == 0])
```

```python
stats_dados_faltantes(df=df[df['default'] == 1])
```

Como a proporção de dados faltantes é muito próxima tanto para adimplentes quanto para inadimplentes podemos ter a segurança de que, por exemplo, retirar essas linhas de nossos dados não afetará ainda mais a questão de desbalanceamento dos dados.

## 4\. Transformação e limpeza de dados

Agora que conhecemos melhor a natureza do nosso conjunto de dados, vamos conduzir uma atividade conhecida como *data wrangling* que consiste na transformação e limpeza dos dados do conjunto para que possam ser melhor analisados. Em especial, vamos remover:

### **4.1. Correção de schema** 

Na etapa de exploração, notamos que as colunas **limite_credito** e **valor_transacoes_12m** estavam sendo interpretadas como colunas categóricas (`dtype = object`).

```python
df[['limite_credito', 'valor_transacoes_12m']].dtypes
```

```python
df[['limite_credito', 'valor_transacoes_12m']].head(n=5)
```

Vamos criar uma função `lambda` para limpar os dados. Mas antes, vamos testar sua aplicação através do método funcional `map`:

```python
fn = lambda valor: float(valor.replace(".", "").replace(",", ".")) # trasforma em float e coloca no modelo correto de float.

valores_originais = ['12.691,51', '8.256,96', '3.418,56', '3.313,03', '4.716,22']
valores_limpos = list(map(fn, valores_originais))

print(valores_originais)
print(valores_limpos)
```

Com a função `lambda` de limpeza pronta, basta aplica-la nas colunas de interesse.

```python
df['valor_transacoes_12m'] = df['valor_transacoes_12m'].apply(fn)
df['limite_credito'] = df['limite_credito'].apply(fn)
```

### **5.1. Visualizações categóricas** 

Nesta seção, vamos visualizar a relação entre a variável resposta **default** com os atributos categóricos.

```python
df.select_dtypes('object').head(n=5)
```

Primeiro criamos uma função para criar a visualização dos dados categóricos:

```python
def vis_categorica(coluna : str, titulos : list) -> None:
    eixo = 0
    max_y = 0
    max = df.select_dtypes('object').describe()[coluna]['freq'] * 1.1

    figura, eixos = plt.subplots(1,3, figsize=(20, 5), sharex=True)

    for dataframe in [df, df_adimplente, df_inadimplente]:

      df_to_plot = dataframe[coluna].value_counts().to_frame()
      df_to_plot.rename(columns={coluna: 'frequencia_absoluta'}, inplace=True)
      df_to_plot[coluna] = df_to_plot.index
      df_to_plot.sort_values(by=[coluna], inplace=True)
      df_to_plot.sort_values(by=[coluna])

      f = sns.barplot(x=df_to_plot[coluna], y=df_to_plot['frequencia_absoluta'], ax=eixos[eixo])
      f.set(title=titulos[eixo], xlabel=coluna.capitalize(), ylabel='Frequência Absoluta')
      f.set_xticklabels(labels=f.get_xticklabels(), rotation=90)

      _, max_y_f = f.get_ylim()
      max_y = max_y_f if max_y_f > max_y else max_y
      f.set(ylim=(0, max_y))

      eixo += 1

    figura.show()
```

 - Escolaridade

```python
coluna = 'escolaridade'
titulos = ['Escolaridade dos Clientes', 'Escolaridade dos Clientes Adimplentes', 'Escolaridade dos Clientes Inadimplentes']

vis_categorica(coluna, titulos)
```

 - Salário Anual

```python
coluna = 'salario_anual'
titulos = ['Salário Anual dos Clientes', 'Salário Anual dos Clientes Adimplentes', 'Salário Anual dos Clientes Inadimplentes']

vis_categorica(coluna, titulos)
```

- Estado civil

```python
coluna = 'estado_civil'
titulos = ['Estado civil dos Clientes', 'Estado civil dos Clientes Adimplentes', 'Estado civil dos Clientes Inadimplentes']

vis_categorica(coluna, titulos)
```

- Tipo de cartão

```python
coluna = 'tipo_cartao'
titulos = ['Tipo de cartão dos Clientes', 'Tipo de cartão dos Clientes Adimplentes', 'Tipo de cartão dos Clientes Inadimplentes']

vis_categorica(coluna, titulos)
```

Em todas as comparações de variáveis categóricas observamos a mesma distribuição tanto para Adimplentes como para inadimplentes, com excessão da variavel **estado civil** que nos mostrou que existe uma quantidade menor de casados Inadimplente (podendo sugerir que pessoas casadas tem mais facilidade para se manter Adimplente, possivelmente pela junção de seus ganhos), porém isso não nos traz nenhum insight sobre o estado civil implicar na inadimplência, pois como podemos ver no último gráfico a quantidade de casado e solteiros inadimplentes é a mesma.

### **5.2. Visualizações numéricas** 

Nesta seção, vamos visualizar a relação entre a variável resposta **default** com os atributos numéricos.

```python
df.drop(['id', 'default'], axis=1).select_dtypes('number').head(n=5)
```

Novamente, faremos uma função que vai gerar a visualização dos dados numéricos:

```python
def vis_numerico(coluna : str, titulos : list) ->None:
    eixo = 0
    max_y = 0
    figura, eixos = plt.subplots(1,3, figsize=(20, 5), sharex=True)

    for dataframe in [df, df_adimplente, df_inadimplente]:

      f = sns.histplot(x=coluna, data=dataframe, stat='count', ax=eixos[eixo])
      f.set(title=titulos[eixo], xlabel=coluna.capitalize(), ylabel='Frequência Absoluta')

      _, max_y_f = f.get_ylim()
      max_y = max_y_f if max_y_f > max_y else max_y
      f.set(ylim=(0, max_y))

      eixo += 1

    figura.show()
```

 - Quantidade de Transações nos Últimos 12 Meses

```python
coluna = 'qtd_transacoes_12m'
titulos = ['Qtd. de Transações no Último Ano', 'Qtd. de Transações no Último Ano de Adimplentes', 'Qtd. de Transações no Último Ano de Inadimplentes']

vis_numerico(coluna,titulos)
```

 - Valor das Transações nos Últimos 12 Meses

```python
coluna = 'valor_transacoes_12m'
titulos = ['Valor das Transações no Último Ano', 'Valor das Transações no Último Ano de Adimplentes', 'Valor das Transações no Último Ano de Inadimplentes']

vis_numerico(coluna,titulos)
```

Aqui podemos observar um comportamento específico dos inadimplentes, a grande maioria tem seus valores de transação em torno dos 1000 e 3000.

- limite de crédito

```python
coluna = 'limite_credito'
titulos = ['Limite de crédito', 'Limite de crédito de Adimplentes', 'Limite de crédito de Inadimplentes']

vis_numerico(coluna,titulos)
```

- Meses de relacionamento

```python
coluna = 'meses_de_relacionamento'
titulos = ['Meses de relacionamento', 'Meses de relacionamento de Adimplentes', 'Meses de relacionamento de Inadimplentes']

vis_numerico(coluna,titulos)
```

- Dependentes

```python
coluna = 'dependentes'
titulos = ['Dependentes', 'Dependentes de Adimplentes', 'Dependentes de Inadimplentes']

vis_numerico(coluna,titulos)
```

- Iterações 12m

```python
coluna = 'iteracoes_12m'
titulos = ['Iterações no Último Ano', 'Iterações no Último Ano de Adimplentes', 'Iterações no Último Ano de Inadimplentes']

vis_numerico(coluna,titulos)
```

De todas as variáveis numéricas que observamos, tivemos diferenças substânciais entre adimplentes e inadimplentes apenas em:  

Valor de Transações nos Últimos 12 Meses  

Quantidade de Transações nos Últimos 12 Meses  

Esperamos instintivamente que essas duas variáveis tenham uma correlação. Vamos então comparar as duas em um mesmo gráfico: 


 - Valor de Transações nos Últimos 12 Meses x Quantidade de Transações nos Últimos 12 Meses

```python
f = sns.relplot(x='valor_transacoes_12m', y='qtd_transacoes_12m', data=df, hue='default')
_ = f.set(
    title='Relação entre Valor e Quantidade de Transações no Último Ano', 
    xlabel='Valor das Transações no Último Ano', 
    ylabel='Quantidade das Transações no Último Ano'
  )
```

Com esse gráfico podemos tirar alguns insights:  
Os clientes que possuem quantidade de transações > 80 **e** Valor das transações > 12000 são clientes que o banco não precisa se preocupar, pois em nossos dados não houve 1 exemplo dentro dessa área que seja inadimplente.  
Agora sobre a inadimplência, podemos ver que esses clientes estão definidos em áreas bem particulares:  
- qtd. de transações entre 60 e 80 **e** valor de transação entre 7500 e 10000
- qtd. de transações entre 0 e 60 **e** valor de transação entre 0 e 3000
